```python
import tkinter as tk
from tkinter import ttk

class SoW_App:
    def __init__(self, root):
        self.root = root
        self.root.geometry('1000x600')  # 창 크기를 넓게 조절

        self.create_widgets()

    def create_widgets(self):
        self.buttons_data = [
        "Color using JUNIOR", 
        "Clarity using JUNIOR", 
        "Cake appearance (lyo)", 
        "pH", 
        "Osmolality", 
        "Visible particle", 
        "sub-visible particle (HIAC)", 
        "sub-visible particle (HORIZON)", 
        "sub-visible particle (MFI)", 
        "A280", 
        "SE-HPLC", 
        "IEX-HPLC", 
        "icIEF", 
        "CE-SDS (R) ", 
        "CE-SDS (NR)", 
        "UNCLE", 
        "HUNKY", 
        "DLS", 
        "DSC", 
        "Hyper DSC", 
        "RP-HPLC", 
        "Viscosity", 
        "OD600", 
        "Reconstitution time (lyo)", 
        "Moisture content (lyo)", 
        "FDM (lyo)"
        ]


        for i, data in enumerate(self.buttons_data):
            btn = tk.Button(self.root, text=data, command=lambda data=data: self.add_data(data))
            btn.grid(row=i, column=0, sticky='ew')

        # 데이터를 표시할 Listbox 초기화
        self.data_listbox = tk.Listbox(self.root, selectmode='extended', width=50)
        self.data_listbox.grid(row=0, column=1, rowspan=40, sticky='nsew')

        # 데이터 삭제 버튼, 3번째 컬럼에 위치
        self.delete_btn = tk.Button(self.root, text="Delete selected Data", command=self.delete_data)
        self.delete_btn.grid(row=0, column=2, sticky='nw')

    def add_data(self, data):
        # 선택된 데이터를 Listbox에 추가
        self.data_listbox.insert(tk.END, data)

    def delete_data(self):
        # 선택된 항목(들)을 삭제
        selected_indices = self.data_listbox.curselection()
        if selected_indices:
            # 선택된 항목이 있을 때, 선택된 모든 항목을 뒤에서부터 삭제
            for i in reversed(selected_indices):
                self.data_listbox.delete(i)
        else:
            # 선택된 항목이 없을 때, 마지막 항목 삭제
            self.data_listbox.delete(tk.END)

if __name__ == "__main__":
    root = tk.Tk()
    app = SoW_App(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

class SoW_App:
    def __init__(self, root):
        self.root = root
        self.root.geometry('1200x600')  # 창의 초기 크기 설정
        self.current_buttons = []  # 현재 생성된 버튼들을 추적하기 위한 리스트

        self.create_dropdown()  # 드롭다운 메뉴 생성
        self.create_button_area()  # 버튼 영역 생성
        self.create_data_display_area()  # 데이터 표시 영역(스크롤바 포함) 생성

    def create_dropdown(self):
        # 드롭다운 메뉴 옵션
        self.options = ['Process', 'Activity', 'Document']
        # 드롭다운 메뉴를 위한 StringVar
        self.combo_var = tk.StringVar()
        # Combobox 생성 및 설정
        self.dropdown = ttk.Combobox(self.root, textvariable=self.combo_var, values=self.options, state='readonly')
        self.dropdown.grid(row=0, column=0, sticky='ew')
        self.dropdown.bind('<<ComboboxSelected>>', self.update_buttons)  # 선택 이벤트 바인딩

    def create_button_area(self):
        # 버튼을 담을 프레임 생성
        self.button_frame = tk.Frame(self.root)
        self.button_frame.grid(row=1, column=0, sticky='nsew')

    def create_data_display_area(self):
        # 데이터 표시를 위한 Listbox 생성 및 위치 설정
        self.data_listbox = tk.Listbox(self.root, selectmode='extended', width=50, height=25)
        self.data_listbox.grid(row=0, column=1, rowspan=40, sticky='nsew', padx=10, pady=5)

        # 스크롤바 생성 및 Listbox에 연결
        self.scrollbar = tk.Scrollbar(self.root, command=self.data_listbox.yview)
        self.scrollbar.grid(row=0, column=2, rowspan=40, sticky='nsew', padx=(0, 20))
        self.data_listbox.configure(yscrollcommand=self.scrollbar.set)

        # 삭제 및 Export 버튼 생성 및 위치 설정 (위쪽으로 정렬)
        self.delete_btn = tk.Button(self.root, text="Delete selected Data", command=self.delete_data)
        self.delete_btn.grid(row=1, column=3, sticky='ew', padx=20)

        self.export_btn = tk.Button(self.root, text="Export Data", command=self.export_data)
        self.export_btn.grid(row=2, column=3, sticky='ew', padx=20)

    def update_buttons(self, event):
        # 현재 버튼들을 모두 제거하고 새로운 버튼들 생성
        for btn in self.current_buttons:
            btn.destroy()
        self.current_buttons.clear()

        # 선택된 드롭다운 메뉴에 따라 버튼 데이터 설정 및 버튼 생성
        selection = self.combo_var.get()
        buttons_data = self.get_buttons_data(selection)
        for data in buttons_data:
            btn = tk.Button(self.button_frame, text=data, command=lambda data=data: self.add_data(data))
            btn.pack()
            self.current_buttons.append(btn)

    def get_buttons_data(self, selection):
        # 드롭다운 선택에 따른 버튼 데이터 반환
        data = {
            'Process': [
        "Color using JUNIOR", 
        "Clarity using JUNIOR", 
        "Cake appearance (lyo)", 
        "pH", 
        "Osmolality", 
        "Visible particle", 
        "sub-visible particle (HIAC)", 
        "sub-visible particle (HORIZON)", 
        "sub-visible particle (MFI)", 
        "A280", 
        "SE-HPLC", 
        "IEX-HPLC", 
        "icIEF", 
        "CE-SDS (R) ", 
        "CE-SDS (NR)", 
        "UNCLE", 
        "HUNKY", 
        "DLS", 
        "DSC", 
        "Hyper DSC", 
        "RP-HPLC", 
        "Viscosity", 
        "OD600", 
        "Reconstitution time (lyo)", 
        "Moisture content (lyo)", 
        "FDM (lyo)"
        ],
            'Activity': ["Activity 1", "Activity 2", "Activity 3"],
            'Document': ["Document 1", "Document 2", "Document 3"]
        }
        return data.get(selection, [])

    def add_data(self, data):
        # 선택된 데이터를 Listbox에 추가
        self.data_listbox.insert(tk.END, data)

    def delete_data(self):
        # Listbox에서 선택된 항목(들) 삭제
        selected_indices = self.data_listbox.curselection()
        for i in reversed(selected_indices):
            self.data_listbox.delete(i)

    def export_data(self):
        # 새 창을 열고 Listbox 내용을 텍스트로 출력
        export_window = tk.Toplevel(self.root)
        export_window.title("Exported Data")
        export_window.geometry("400x600")
        text_area = tk.Text(export_window, wrap="word")
        text_area.pack(padx=10, pady=10, fill=tk.BOTH, expand=True)
        for item in self.data_listbox.get(0, tk.END):
            text_area.insert(tk.END, item + "\n")
        text_area.config(state=tk.DISABLED)  # 편집 방지

if __name__ == "__main__":
    root = tk.Tk()
    app = SoW_App(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

class SoW_App:
    def __init__(self, root):
        self.root = root
        self.root.geometry('1200x600')  # 창의 크기 설정

        self.data = {
            'Process': ["Process 1", "Process 2", "Process 3"],
            'Activity': ["Activity 1", "Activity 2", "Activity 3"],
            'Document': ["Document 1", "Document 2", "Document 3"]
        }  # 데이터 설정

        self.create_scrollable_area()  # 스크롤 가능한 영역 생성
        self.populate_buttons()  # 버튼 생성

    def create_scrollable_area(self):
        # 스크롤바와 함께 스크롤 가능한 영역 생성
        self.canvas = tk.Canvas(self.root)
        self.scrollbar = ttk.Scrollbar(self.root, orient="vertical", command=self.canvas.yview)
        self.scrollable_frame = ttk.Frame(self.canvas)

        self.scrollable_frame.bind(
            "<Configure>",
            lambda e: self.canvas.configure(
                scrollregion=self.canvas.bbox("all")
            )
        )

        self.canvas.create_window((0, 0), window=self.scrollable_frame, anchor="nw")
        self.canvas.configure(yscrollcommand=self.scrollbar.set)

        self.canvas.grid(row=0, column=0, sticky="nsew")
        self.scrollbar.grid(row=0, column=1, sticky='ns')

    def populate_buttons(self):
        # 큰 제목과 하위 버튼들 생성
        for key, values in self.data.items():
            label = tk.Label(self.scrollable_frame, text=key, font=("Helvetica", 16))
            label.pack(pady=10)
            for value in values:
                btn = tk.Button(self.scrollable_frame, text=value, command=lambda value=value: self.button_action(value))
                btn.pack()

    def button_action(self, value):
        # 버튼 클릭 시 실행할 액션
        print(f"Button clicked: {value}")

if __name__ == "__main__":
    root = tk.Tk()
    app = SoW_App(root)
    root.mainloop()
```
